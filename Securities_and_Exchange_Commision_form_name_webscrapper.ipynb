This is a python script that was designed to count and create a list of the number of SEC filing forms. After running this script, the output will be a CSV file that contains the complete list of SEC filing forms.

```python
import requests
import bs4
import csv
```

```python
webpage = 'https://www.sec.gov/forms'
server_response = requests.get(webpage)
print(server_response)
```

```python
soup = bs4.BeautifulSoup(server_response.text, features="html.parser")
```

```python
empty_list = []
for link in soup.find_all('a'):
    empty_list.append(link.get('href'))
```

```python
len(empty_list)
h_list = []
for item in empty_list:
    if type(item) == str:
        h_list.append(item)
h_list
```

```python
forms = []
for thing in h_list:
    if thing[0:11] == '/files/form':
        forms.append(thing)
    else:
        pass
len(forms)
```

```python
with open('sec_forms.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerows([forms])
```
